##### We use CTGAN (Conditional Tabular GAN) and why?
###### Vanilla GAN struggle with working well with imbalanced, mixed discrete + continuous features. CTGAN can solve this.
###### CTGAN is used to generate synthetic tabular data (like CSV datasets) that mimics the statistical properties of real data. It’s part of the SDV [(Synthetic Data Vault)](https://github.com/sdv-dev) ecosystem.

#### A Robust Pipeline for Synthetic Data Generation with CTGAN (https://github.com/sdv-dev/CTGAN)
##### An enhanced Python script featuring data preprocessing, quality evaluation and dataset generation.

```python
import pandas as pd
from ctgan import CTGAN
from sdv.evaluation.single_table import evaluate_quality

# --- 1. Configuration & Best Practices ---

# Use variables for paths and parameters for easier management.

DATA_PATH = "data/synthetic_clinical_dataset.csv"

# Epoch Count
EPOCHS = 3000

# Number of rows to generate
n=1000
# --- 2. Data Loading & Preprocessing ---
try:
    # We load the REAL data to train the model correctly. (REAL data here is the synthetic data we generated in the first place)
    real_data = pd.read_csv(DATA_PATH)
    print(f"Successfully loaded data from {DATA_PATH}. Shape: {real_data.shape}")
except FileNotFoundError:
    print(f"Error: The file was not found at {DATA_PATH}.")
    print("Please ensure the path is correct and the script is run from the right directory.")
    exit()

# Data Preprocessing (Missing Values)
if real_data.isnull().sum().any():
    print(f"Missing values found. Dropping {real_data.isnull().any(axis=1).sum()} rows with NaNs.")
    real_data.dropna(inplace=True)
    print(f"Data shape after dropping NaNs: {real_data.shape}")

# Defining columns that are categorical. (needed for ctgan library)
discrete_columns = [
    'menopausal_status',
    'parity',
    'family_history',
    'mht_use',
    'brca_status',
    'diagnosis_label'
]

# Column Names Mismatch
# A simple check to ensure all specified discrete columns exist in the DataFrame.
missing_cols = [col for col in discrete_columns if col not in real_data.columns]
if missing_cols:
    print(f"Error: The following discrete columns are not in the dataset: {missing_cols}")
    exit()

# --- 3. Model Training ---

# Initialize the CTGAN model with declared parameters.
ctgan = CTGAN(
    epochs=EPOCHS,
    verbose=True # Set to True to see training progress.
)

print("\nStarting CTGAN model training...")
ctgan.fit(real_data, discrete_columns)
print("Training complete.")

# --- 4. Generate & Save Synthetic Data ---

print("\nGenerating synthetic data...")
# Create synthetic data with the n number of rows
synthetic_data = ctgan.sample(n)
print("Synthetic data generated successfully.")

# Save the data for later use.
synthetic_data.to_csv("data/ctGAN_synthetic_data_generated.csv", index=False)
print("Synthetic data saved to 'data/synthetic_data_generated.csv'")


# --- 5. Evaluation ---

# Evaluation of synthetic data quality.
print("\nEvaluating the quality of the synthetic data...")

from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
metadata.detect_from_dataframe(data=real_data)

# Pass the metadata object to the evaluation function ---
print("\nEvaluating the quality of the synthetic data...")
quality_report = evaluate_quality(
    real_data=real_data,
    synthetic_data=synthetic_data,
    metadata=metadata
)

# overall score (0..1)
overall = quality_report.get_score()
print("Overall quality score:", overall)

# list available properties
print("Available properties:", quality_report.get_properties())

# get details for a valid property
details = quality_report.get_details('Column Shapes')   # valid for single-table
print(details)
```

In many of experiments, the generator loss started positive and eventually stabilized at a negative value while the discriminator loss oscillated around 0. This is shown below for a variety of different datasets.

```python
import numpy as np
import pandas as pd


df=pd.read_csv("data/ctGAN_synthetic_data_generated.csv")

def pct(x):
    try: return 100*float(x)
    except: return np.nan

def report(desc, ok, val):
    status = "PASS" if ok else "FAIL"
    # Pretty-print tuples and scalars
    if isinstance(val, tuple):
        v = ", ".join(f"{x:.3f}" if isinstance(x, (int,float,np.floating)) else str(x) for x in val)
    elif isinstance(val, (int,float,np.floating)):
        v = f"{val:.3f}"
    else:
        v = str(val)
    print(f"[{status}] {desc:<60} -> {v}")

controls = df[df["diagnosis_label"] == 0]
cases    = df[df["diagnosis_label"] == 1]

checks = []

# --- Cohort composition ---
prev = df["diagnosis_label"].mean()
checks.append(("Cancer fraction ≥ 30%", prev >= 0.30, prev))

# --- BRCA prevalence ---
brca_case_rate = cases["brca_status"].mean() if len(cases) else np.nan
checks.append(("BRCA+ among cases in [12%, 18%]", 0.12 <= brca_case_rate <= 0.18, brca_case_rate))

# --- CA-125 behavior ---
healthy_high35 = (controls["ca125"] > 35).mean() if len(controls) else np.nan
case_high200   = (cases["ca125"] > 200).mean() if len(cases) else np.nan
checks.append(("Healthy CA-125 >35 U/mL ≈ 1–3%", 0.01 <= healthy_high35 <= 0.03, healthy_high35))
checks.append(("Cancer CA-125 >200 U/mL ≥ 35%", case_high200 >= 0.35, case_high200))

# --- Tumor size ---
median_size_cases = cases["tumor_size_cm"].median() if len(cases) else np.nan
nz_ctrl = controls.loc[controls["tumor_size_cm"] > 0, "tumor_size_cm"]
median_size_ctrl_nonzero = nz_ctrl.median() if len(nz_ctrl) else np.nan
checks.append(("Median tumor size (cases) in [8, 11] cm", 8.0 <= median_size_cases <= 11.0, median_size_cases))
if pd.notna(median_size_ctrl_nonzero):
    checks.append(("Median benign cyst (controls) in [2, 4] cm", 2.0 <= median_size_ctrl_nonzero <= 4.0, median_size_ctrl_nonzero))

# Zero-mass proportion in controls (expect high; ~60–80%)
zero_mass_rate = (controls["tumor_size_cm"] == 0).mean() if len(controls) else np.nan
checks.append(("Controls: zero-mass proportion in [0.6, 0.8]", 0.60 <= zero_mass_rate <= 0.80, zero_mass_rate))

# --- Ultrasound risk sanity ---
mean_risk_cases = cases["ultrasound_risk_score"].mean() if len(cases) else np.nan
mean_risk_ctrls = controls["ultrasound_risk_score"].mean() if len(controls) else np.nan
checks.append(("Mean risk (cases) > Mean risk (controls)", mean_risk_cases > mean_risk_ctrls, (mean_risk_cases, mean_risk_ctrls)))

# Controls w/ no mass should be very low risk (use 95th percentile to avoid mean skew)
no_mass = controls.loc[controls["tumor_size_cm"] == 0, "ultrasound_risk_score"]
q95 = no_mass.quantile(0.95) if len(no_mass) else np.nan
checks.append(("Controls no-mass: 95th percentile risk ≤ 0.08", q95 <= 0.08, q95))

# --- Menopause / MHT checks ---
p_post_55 = (df.loc[df["age"] >= 55, "menopausal_status"].mean()
             if (df["age"] >= 55).any() else np.nan)
checks.append(("P(postmenopause | age ≥ 55) in [0.80, 0.90]", 0.80 <= p_post_55 <= 0.90, p_post_55))

mht_post_ctrl = controls.loc[controls["menopausal_status"] == 1, "mht_use"].mean() if (controls["menopausal_status"] == 1).any() else np.nan
mht_post_case = cases.loc[cases["menopausal_status"] == 1, "mht_use"].mean() if (cases["menopausal_status"] == 1).any() else np.nan
checks.append(("MHT among post-menopause (controls) ≈ 0.20±0.05", 0.15 <= mht_post_ctrl <= 0.25, mht_post_ctrl))
checks.append(("MHT among post-menopause (cases) ≈ 0.25±0.05", 0.20 <= mht_post_case <= 0.30, mht_post_case))

# --- Hemodilution sanity (controls only): BMI vs CA-125 should be negative ---
rho_bmi_ca125_ctrl = controls[["bmi","ca125"]].corr(method="spearman").iloc[0,1] if len(controls) else np.nan
checks.append(("Controls: Spearman(BMI, CA-125) < 0", rho_bmi_ca125_ctrl < 0, rho_bmi_ca125_ctrl))

# --- Within-cases correlations (intended behavior) ---
def safe_corr(df_sub, a, b, method="spearman"):
    if len(df_sub) < 3 or df_sub[[a,b]].dropna().shape[0] < 3:
        return np.nan
    return df_sub[[a,b]].corr(method=method).iloc[0,1]

rho_size_risk_case  = safe_corr(cases, "tumor_size_cm", "ultrasound_risk_score")
rho_ca_risk_case    = safe_corr(cases, "ca125", "ultrasound_risk_score")
rho_size_ca_case    = safe_corr(cases, "tumor_size_cm", "ca125")

checks.append(("Cases: corr(size, risk) > 0.5", rho_size_risk_case > 0.5, rho_size_risk_case))
checks.append(("Cases: corr(CA-125, risk) in [0.3, 0.5]", 0.3 <= rho_ca_risk_case <= 0.5, rho_ca_risk_case))
checks.append(("Cases: corr(size, CA-125) > 0", rho_size_ca_case > 0.0, rho_size_ca_case))

# --- BRCA & age relationship (you already had this; keeping it) ---
med_age_brca1 = cases.loc[cases["brca_status"]==1, "age"].median() if (cases["brca_status"]==1).any() else np.nan
med_age_brca0 = cases.loc[cases["brca_status"]==0, "age"].median() if (cases["brca_status"]==0).any() else np.nan
checks.append(("BRCA+ cases skew younger (median)", med_age_brca1 < med_age_brca0, (med_age_brca1, med_age_brca0)))

# --- Hard bounds sanity (no impossible values) ---
ok_bounds = (
    (df["bmi"].between(16, 45).all()) and
    (df["age"].between(18, 90).all()) and
    (df["ca125"].between(2, 2000).all()) and
    (df.loc[df["diagnosis_label"]==1, "tumor_size_cm"].between(0.5, 20).all()) and
    (df.loc[df["diagnosis_label"]==0, "tumor_size_cm"].ge(0).all())
)
checks.append(("All features within specified bounds", ok_bounds, ok_bounds))

# --- Print report ---
print("Plausibility checks:")
for desc, ok, val in checks:
    report(desc, ok, val)
```

### Switching to CTGANSynthesizer (a straightforward, integrated way to generate synthetic data)

#### More importantly, with CTGANSynthesizer from sdv, we can manage constraints to an extent which is not supported when using the standalone CTGAN library. Constraints are provided only through the SDV (Synthetic Data Vault) interface, not in the bare CTGAN.

```python
import numpy as np
import pandas as pd

# ---------------- Settings (same as your draft) ----------------
DATA_PATH = "data/synthetic_clinical_dataset.csv"
OUT_PATH = "data/ctGANSynthesizer_synthetic_data.csv"
EPOCHS = 3000
n = 1000
RANDOM_STATE = 42

# ---------------- Load Real Data ----------------
real_data = pd.read_csv(DATA_PATH)
print(f"Loaded real data: {real_data.shape}")

discrete_columns = [
    "menopausal_status",
    "parity",
    "family_history",
    "mht_use",
    "brca_status",
    "diagnosis_label",
]
missing = [c for c in discrete_columns if c not in real_data.columns]
if missing:
    raise ValueError(f"Missing discrete columns in real data: {missing}")

# ---------------- SDV imports ----------------
from sdv.metadata import SingleTableMetadata
from sdv.single_table import CTGANSynthesizer
from sdv.constraints.base import Constraint as BaseConstraint
from sdv.evaluation.single_table import evaluate_quality

# ---------------- Build metadata ----------------
metadata = SingleTableMetadata()
metadata.detect_from_dataframe(real_data)
# declare discrete columns as categorical to help the synthesizer
metadata.update_columns(column_names=discrete_columns, sdtype="categorical")

# ---------------- Programmed single-table constraints (row-level logic) ----------------
# All constraints subclass BaseConstraint and set _is_single_table = True so SDV will accept them.

class TumorSizeByDiagnosisConstraint(BaseConstraint):
    """
    Row-level rule:
      - if diagnosis_label == 1 (case) -> tumor_size_cm >= 0.5
      - if diagnosis_label == 0 (control) -> tumor_size_cm >= 0.0
    This enforces the tumor-size floor per diagnosis (your post-processing did the same).
    """
    _is_single_table = True

    def __init__(self, tumor_col="tumor_size_cm", diag_col="diagnosis_label"):
        super().__init__()
        self.tcol = tumor_col
        self.dcol = diag_col
        self.columns = [self.tcol, self.dcol]
        self.column_names = [self.tcol, self.dcol]

    def is_valid(self, table: pd.DataFrame) -> pd.Series:
        if self.tcol not in table.columns or self.dcol not in table.columns:
            # If columns missing, treat rows as invalid (SDV expects real data to be valid)
            return pd.Series([False]*len(table), index=table.index)
        t = table[self.tcol].astype(float)
        d = table[self.dcol].astype(float)
        valid_case = (d == 1) & (t >= 0.5)
        valid_ctrl = (d == 0) & (t >= 0.0)
        return (valid_case | valid_ctrl).fillna(False)

    def transform(self, table: pd.DataFrame) -> pd.DataFrame:
        # no-op transform to let SDV choose transform handling strategy
        return table.copy()

    def reverse_transform(self, table: pd.DataFrame) -> pd.DataFrame:
        return table.copy()

    def get_updated_metadata(self, metadata):
        return metadata

    # Accept whatever SDV passes; ignore extra kwargs if not needed
    def fit(self, data=None, metadata=None, **kwargs):
        """
        SDV will call constraint.fit(data=data, metadata=metadata).
        Implement any fitting logic here (e.g., compute lookup stats), or just mark fitted.
        """
        # Example: we don't need to learn parameters, so just mark fitted.
        self._fitted = True
        return self


class NoMassControlLowRiskConstraint(BaseConstraint):
    """
    Row-level rule:
      - if diagnosis_label == 0 and tumor_size_cm == 0 -> ultrasound_risk_score <= 0.079
    This mirrors your post-processing clamp for controls with no mass.
    """
    _is_single_table = True

    def __init__(self, tumor_col="tumor_size_cm", diag_col="diagnosis_label", risk_col="ultrasound_risk_score"):
        super().__init__()
        self.tcol = tumor_col
        self.dcol = diag_col
        self.rcol = risk_col
        self.columns = [self.tcol, self.dcol, self.rcol]
        self.column_names = [self.tcol, self.dcol, self.rcol]

    def is_valid(self, table: pd.DataFrame) -> pd.Series:
        # If any required column missing, return all False (SDV requires real data satisfy constraints)
        if any(c not in table.columns for c in (self.tcol, self.dcol, self.rcol)):
            return pd.Series([False]*len(table), index=table.index)

        t = table[self.tcol].astype(float)
        d = table[self.dcol].astype(int)
        r = table[self.rcol].astype(float)

        # rows that are "no-mass controls" must have risk <= 0.079
        mask_no_mass_ctrl = (d == 0) & (t == 0)
        # valid if either not a no-mass-control, or risk <= 0.079 for those rows
        valid = (~mask_no_mass_ctrl) | (r <= 0.079)
        return valid.fillna(False)

    def transform(self, table: pd.DataFrame) -> pd.DataFrame:
        # no-op
        return table.copy()

    def reverse_transform(self, table: pd.DataFrame) -> pd.DataFrame:
        return table.copy()
    def get_updated_metadata(self, metadata):
        return metadata

    # Accept whatever SDV passes; ignore extra kwargs if not needed
    def fit(self, data=None, metadata=None, **kwargs):
        """
        SDV will call constraint.fit(data=data, metadata=metadata).
        Implement any fitting logic here (e.g., compute lookup stats), or just mark fitted.
        """
        # Example: we don't need to learn parameters, so just mark fitted.
        self._fitted = True
        return self

class UltrasoundRiskBoundsConstraint(BaseConstraint):
    """
    Row-level rule:
      - ultrasound_risk_score must be within [0.0, 0.99]
    (helps guard against outliers that violate post-processing.)
    """
    _is_single_table = True

    def __init__(self, risk_col="ultrasound_risk_score"):
        super().__init__()
        self.rcol = risk_col
        self.columns = [self.rcol]
        self.column_names = [self.rcol]

    def is_valid(self, table: pd.DataFrame) -> pd.Series:
        if self.rcol not in table.columns:
            return pd.Series([False]*len(table), index=table.index)
        r = pd.to_numeric(table[self.rcol], errors="coerce")
        return r.between(0.0, 0.99).fillna(False)

    def transform(self, table: pd.DataFrame) -> pd.DataFrame:
        return table.copy()

    def reverse_transform(self, table: pd.DataFrame) -> pd.DataFrame:
        return table.copy()
    def get_updated_metadata(self, metadata):
        return metadata

    # Accept whatever SDV passes; ignore extra kwargs if not needed
    def fit(self, data=None, metadata=None, **kwargs):
        """
        SDV will call constraint.fit(data=data, metadata=metadata).
        Implement any fitting logic here (e.g., compute lookup stats), or just mark fitted.
        """
        # Example: we don't need to learn parameters, so just mark fitted.
        self._fitted = True
        return self

class IntegerBinaryColumnsConstraint(BaseConstraint):
    """
    Row-level rule:
      - ensure listed columns are integer/binary (0/1)
      This constraint marks rows valid only if those columns are 0/1 integers.
    """
    _is_single_table = True

    def __init__(self, cols=None):
        super().__init__()
        self.cols = cols or ["menopausal_status", "family_history", "mht_use", "brca_status", "diagnosis_label", "parity"]
        self.columns = self.cols.copy()
        self.column_names = self.cols.copy()

    def is_valid(self, table: pd.DataFrame) -> pd.Series:
        # If a column is missing, treat that column as invalid (SDV expects real data to comply)
        for c in self.cols:
            if c not in table.columns:
                return pd.Series([False]*len(table), index=table.index)
        # For parity allow 0..8 integers; for binary columns allow 0 or 1
        valid = pd.Series(True, index=table.index)
        for c in self.cols:
            if c == "parity":
                valid &= table[c].apply(lambda x: (pd.notna(x) and float(x).is_integer() and 0 <= int(x) <= 8))
            else:
                valid &= table[c].apply(lambda x: (pd.notna(x) and float(x) in (0.0, 1.0)))
        return valid.fillna(False)

    def transform(self, table: pd.DataFrame) -> pd.DataFrame:
        return table.copy()

    def reverse_transform(self, table: pd.DataFrame) -> pd.DataFrame:
        return table.copy()

    def get_updated_metadata(self, metadata):
        return metadata

    # Accept whatever SDV passes; ignore extra kwargs if not needed
    def fit(self, data=None, metadata=None, **kwargs):
        """
        SDV will call constraint.fit(data=data, metadata=metadata).
        Implement any fitting logic here (e.g., compute lookup stats), or just mark fitted.
        """
        # Example: we don't need to learn parameters, so just mark fitted.
        self._fitted = True
        return self

# ---------------- Create constraint list ----------------
constraints = [
    TumorSizeByDiagnosisConstraint(tumor_col="tumor_size_cm", diag_col="diagnosis_label"),
    NoMassControlLowRiskConstraint(tumor_col="tumor_size_cm", diag_col="diagnosis_label", risk_col="ultrasound_risk_score"),
    UltrasoundRiskBoundsConstraint(risk_col="ultrasound_risk_score"),
    IntegerBinaryColumnsConstraint(cols=["menopausal_status", "family_history", "mht_use", "brca_status", "diagnosis_label", "parity"]),
]

# ---------------- Instantiate CTGAN with min/max enforcement ----------------
# enforce_min_max_values ensures numeric columns are clipped to observed min/max during generation.
synth = CTGANSynthesizer(
    metadata=metadata,
    epochs=EPOCHS,
    verbose=True,
    enforce_min_max_values=True,
)
print("Initialized CTGANSynthesizer with enforce_min_max_values=True")

# Add programmed constraints (SDV v1.26.0: use add_constraints after instantiation)
synth.add_constraints(constraints)
print(f"Added {len(constraints)} programmed single-table constraints (row-level).")

# ---------------- Fit and Sample ----------------
print("Fitting synthesizer (this may take a while)...")
synth.fit(real_data)
print("Fit complete.")

print(f"Sampling {n} synthetic rows...")
synthetic_data = synth.sample(num_rows=n)
print("Sampling complete. Synthetic shape:", synthetic_data.shape)

# Save synthetic output
synthetic_data.to_csv(OUT_PATH, index=False)
print(f"Saved synthetic data to '{OUT_PATH}'")

# ---------------- 5. Evaluation (your same block) ----------------
print("\nEvaluating the quality of the synthetic data...")

quality_report = evaluate_quality(
    real_data=real_data,
    synthetic_data=synthetic_data,
    metadata=metadata,
    verbose=True
)

overall = quality_report.get_score()
print("Overall quality score:", overall)

print("Available properties:", quality_report.get_properties())

try:
    details = quality_report.get_details('Column Shapes')
    print("\nColumn Shapes details (first rows):")
    print(details.head())
except Exception as e:
    print("Could not fetch 'Column Shapes' details:", e)

print("\nDone.")
```

#### As can be seen above ⬆️, 6 (instead of previuos 9) FAIL out of 18 metrics. Solving them as follows.

## Why some checks still failed even with constraints in place?
#### Because CTGAN learns patterns, not hard rules. Even with Range constraints, the model won’t enforce distributional expectations (e.g., “CA-125 >200 in ≥35% of cases”) because that’s not a strict bound, it’s a statistical prior.
#### Secondly, discrete distributions may drift. Even though we round and cast, CTGAN may under/over-sample categories (like BRCA carriers or postmenopausal MHT use).

#### Post-ctGAN: Enforcing strict numeric bounds, deterministic post-sample repair and tuning pipeline to meet epidemiologic and correlation-based plausibility checks.

```python
import numpy as np
import pandas as pd

synthetic_data = pd.read_csv("data/ctGANSynthesizer_synthetic_data.csv")

def enforce_constraints(synthetic_data: pd.DataFrame, seed: int = 42):
    """
    Post-process synthetic_data (pandas.DataFrame) to satisfy plausibility checks.
    Returns the modified DataFrame (copy).
    """
    rng = np.random.default_rng(seed)
    df = synthetic_data.copy()

    # ---------- Helper shortcuts ----------
    def sample_empirical(series, n):
        s = series.dropna()
        if s.empty:
            return np.array([np.nan] * n)
        return s.sample(n=n, replace=True, random_state=seed).to_numpy()

    # ---------- 0) Hard bounds & types (basic sanitization) ----------
    df['age'] = pd.to_numeric(df.get('age'), errors='coerce').fillna(18).astype(int)
    df['age'] = df['age'].clip(18, 90).astype(int)

    df['bmi'] = pd.to_numeric(df.get('bmi'), errors='coerce').fillna(25.0).astype(float)
    df['bmi'] = df['bmi'].clip(16.0, 45.0).round(2)

    df['ca125'] = pd.to_numeric(df.get('ca125'), errors='coerce').fillna(10.0).astype(float)
    df['ca125'] = df['ca125'].clip(2.0, 2000.0)

    df['tumor_size_cm'] = pd.to_numeric(df.get('tumor_size_cm'), errors='coerce').fillna(0.0).astype(float)

    for col in ['menopausal_status', 'family_history', 'mht_use', 'brca_status', 'diagnosis_label']:
        if col in df.columns:
            df[col] = (pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int) > 0).astype(int)

    if 'parity' in df.columns:
        df['parity'] = pd.to_numeric(df['parity'], errors='coerce').fillna(0).astype(int)
        df['parity'] = df['parity'].clip(0, 8)

    # ---------- 1) Enforce prevalence (Cancer fraction ≥ 30%) ----------
    n = len(df)
    if n == 0:
        return df

    prev = df['diagnosis_label'].mean()
    if prev < 0.30:
        need = int(np.ceil(0.30 * n - df['diagnosis_label'].sum()))
        candidates = df.index[df['diagnosis_label'] == 0].to_numpy()
        if candidates.size > 0:
            to_flip = rng.choice(candidates, size=min(need, candidates.size), replace=False)
            df.loc[to_flip, 'diagnosis_label'] = 1
            existing_cases = df[df['diagnosis_label'] == 1]
            if len(existing_cases) >= 1:
                df.loc[to_flip, 'tumor_size_cm'] = sample_empirical(existing_cases['tumor_size_cm'], len(to_flip))
                df.loc[to_flip, 'ca125'] = sample_empirical(existing_cases['ca125'], len(to_flip))
            else:
                df.loc[to_flip, 'tumor_size_cm'] = rng.uniform(0.5, 12.0, size=len(to_flip))
                df.loc[to_flip, 'ca125'] = rng.uniform(40.0, 250.0, size=len(to_flip))
            if 'family_history' in df.columns:
                fh = df.loc[to_flip, 'family_history'].fillna(0).astype(int).to_numpy()
            else:
                fh = np.zeros(len(to_flip), dtype=int)
            p = np.where(fh == 1, 0.18, 0.12)
            df.loc[to_flip, 'brca_status'] = (rng.random(len(p)) < p).astype(int)
            brca_idx = to_flip[df.loc[to_flip, 'brca_status'] == 1]
            if len(brca_idx) > 0:
                df.loc[brca_idx, 'age'] = np.clip(df.loc[brca_idx, 'age'] - rng.integers(6, 11, size=len(brca_idx)), 18, 90)

    # ---------- 2) BRCA+ among cases in [12%, 18%] ----------
    case_idx = df.index[df['diagnosis_label'] == 1].to_numpy()
    if len(case_idx) >= 1:
        current_rate = df.loc[case_idx, 'brca_status'].mean()
        target_low, target_high, target_mid = 0.12, 0.18, 0.15
        if current_rate < target_low:
            candidates = df.loc[case_idx].index[df.loc[case_idx, 'brca_status'] == 0].to_numpy()
            fh_candidates = [i for i in candidates if ('family_history' in df.columns and df.at[i, 'family_history'] == 1)]
            chosen_pool = np.array(fh_candidates if len(fh_candidates) >= 1 else candidates)
            need = int(np.ceil(target_mid * len(case_idx) - df.loc[case_idx, 'brca_status'].sum()))
            if need > 0 and chosen_pool.size > 0:
                to_set = rng.choice(chosen_pool, size=min(need, chosen_pool.size), replace=False)
                df.loc[to_set, 'brca_status'] = 1
        elif current_rate > target_high:
            pos = df.loc[case_idx].index[df.loc[case_idx, 'brca_status'] == 1].to_numpy()
            need = int(np.ceil(df.loc[case_idx, 'brca_status'].sum() - target_mid * len(case_idx)))
            if need > 0 and pos.size > 0:
                to_unset = rng.choice(pos, size=min(need, pos.size), replace=False)
                df.loc[to_unset, 'brca_status'] = 0

    # ---------- 3) Healthy CA-125 >35 U/mL ≈ 1–3% (controls) - initial pass ----------
    # (kept early pass but final enforcement will be applied later)
    ctrl_idx = df.index[df['diagnosis_label'] == 0].to_numpy()
    if len(ctrl_idx) >= 1:
        n_ctrl = len(ctrl_idx)
        controls = df.loc[ctrl_idx]
        high35_idx = controls[controls['ca125'] > 35].index.to_numpy()
        current_high = len(high35_idx)
        target_min = int(np.ceil(0.01 * n_ctrl))
        target_max = int(np.floor(0.03 * n_ctrl))
        if target_max < target_min:
            target = int(round(0.02 * n_ctrl))
            target_min = target_max = max(0, target)
        if current_high > target_max:
            reduce_n = current_high - target_max
            sel = rng.choice(high35_idx, size=min(reduce_n, len(high35_idx)), replace=False)
            df.loc[sel, 'ca125'] = rng.uniform(20.0, 34.0, size=len(sel))
        elif current_high < target_min:
            low_idx = controls[controls['ca125'] <= 35].index.to_numpy()
            increase_n = target_min - current_high
            if low_idx.size > 0 and increase_n > 0:
                sel = rng.choice(low_idx, size=min(increase_n, low_idx.size), replace=False)
                new_vals = rng.uniform(36.0, 60.0, size=len(sel))
                df.loc[sel, 'ca125'] = np.maximum(df.loc[sel, 'ca125'], new_vals)

    # ---------- 4) Ensure >=35% of cases have CA-125 >200 U/mL ----------
    if len(case_idx) >= 1:
        cases = df.loc[case_idx]
        over200 = cases[cases['ca125'] > 200].index.to_numpy()
        need = int(np.ceil(0.35 * len(case_idx) - len(over200)))
        if need > 0:
            cand = cases[cases['ca125'] <= 200].index.to_numpy()
            if cand.size > 0:
                sel = rng.choice(cand, size=min(need, cand.size), replace=False)
                df.loc[sel, 'ca125'] = rng.uniform(210.0, 600.0, size=len(sel))

    # ---------- 5) Tumor size medians: cases in [8,11], benign cysts (controls nonzero) in [2,4] ----------
    if len(case_idx) >= 3:
        case_sizes = df.loc[case_idx, 'tumor_size_cm'].clip(0.5, 20.0)
        med = case_sizes.median()
        if not (8.0 <= med <= 11.0):
            num_replace = max(1, int(0.15 * len(case_idx)))
            replace_idx = rng.choice(case_idx, size=num_replace, replace=False)
            large_pool = case_sizes[case_sizes > 8.0].dropna()
            if len(large_pool) >= 1:
                df.loc[replace_idx, 'tumor_size_cm'] = sample_empirical(large_pool, len(replace_idx))
            else:
                df.loc[replace_idx, 'tumor_size_cm'] = rng.lognormal(mean=np.log(9.5), sigma=0.5, size=len(replace_idx))
            df.loc[case_idx, 'tumor_size_cm'] = df.loc[case_idx, 'tumor_size_cm'].clip(0.5, 20.0)

    ctrl_nonzero_idx = df.index[(df['diagnosis_label'] == 0) & (df['tumor_size_cm'] > 0)].to_numpy()
    zero_ctrl_idx = df.index[(df['diagnosis_label'] == 0) & (df['tumor_size_cm'] == 0)].to_numpy()
    if len(ctrl_nonzero_idx) >= 1:
        med_ctrl = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].median()
    else:
        med_ctrl = np.nan

    min_nonzero_needed = 3
    if len(ctrl_nonzero_idx) < min_nonzero_needed and len(zero_ctrl_idx) > 0:
        make = min(min_nonzero_needed - len(ctrl_nonzero_idx), len(zero_ctrl_idx))
        make_idx = rng.choice(zero_ctrl_idx, size=make, replace=False)
        df.loc[make_idx, 'tumor_size_cm'] = rng.uniform(2.0, 4.0, size=len(make_idx))
        ctrl_nonzero_idx = df.index[(df['diagnosis_label'] == 0) & (df['tumor_size_cm'] > 0)].to_numpy()
        med_ctrl = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].median()

    if len(ctrl_nonzero_idx) >= 3:
        attempts = 0
        while (pd.isna(med_ctrl) or med_ctrl < 2.0 or med_ctrl > 4.0) and attempts < 20:
            attempts += 1
            med_ctrl = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].median()
            if med_ctrl < 2.0:
                k = len(ctrl_nonzero_idx)
                frac = min(0.2 + 0.05 * attempts, 0.85)
                replace_n = max(1, int(np.ceil(frac * k)))
                smallest = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].sort_values().index.to_numpy()[:replace_n]
                df.loc[smallest, 'tumor_size_cm'] = rng.uniform(2.0, 4.0, size=len(smallest))
            elif med_ctrl > 4.0:
                k = len(ctrl_nonzero_idx)
                frac = min(0.2 + 0.05 * attempts, 0.85)
                replace_n = max(1, int(np.ceil(frac * k)))
                largest = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].sort_values(ascending=False).index.to_numpy()[:replace_n]
                df.loc[largest, 'tumor_size_cm'] = rng.uniform(2.0, 4.0, size=len(largest))
            ctrl_nonzero_idx = df.index[(df['diagnosis_label'] == 0) & (df['tumor_size_cm'] > 0)].to_numpy()
            if len(ctrl_nonzero_idx) == 0:
                med_ctrl = np.nan
                break
            med_ctrl = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].median()

        if not (2.0 <= med_ctrl <= 4.0):
            k = len(ctrl_nonzero_idx)
            if k > 0:
                need_at_or_above = k // 2 + 1
                current_at_or_above = (df.loc[ctrl_nonzero_idx, 'tumor_size_cm'] >= 2.0).sum()
                to_promote = max(0, need_at_or_above - current_at_or_above)
                if to_promote > 0:
                    smallest = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].sort_values().index.to_numpy()[:to_promote]
                    df.loc[smallest, 'tumor_size_cm'] = rng.uniform(2.0, 4.0, size=len(smallest))
            ctrl_nonzero_idx = df.index[(df['diagnosis_label'] == 0) & (df['tumor_size_cm'] > 0)].to_numpy()
            if len(ctrl_nonzero_idx) >= 1:
                med_ctrl = df.loc[ctrl_nonzero_idx, 'tumor_size_cm'].median()

    # ---------- 6) Controls zero-mass proportion in [0.6,0.8] ----------
    if len(ctrl_idx) >= 1:
        zero_mask = (df.loc[ctrl_idx, 'tumor_size_cm'] == 0)
        zero_rate = zero_mask.mean()
        if zero_rate < 0.60:
            cyst_candidates = df.loc[ctrl_idx].loc[df.loc[ctrl_idx, 'tumor_size_cm'] > 0].index.to_numpy()
            need = int(np.ceil(0.60 * len(ctrl_idx) - zero_mask.sum()))
            if cyst_candidates.size > 0 and need > 0:
                flip = rng.choice(cyst_candidates, size=min(need, cyst_candidates.size), replace=False)
                df.loc[flip, 'tumor_size_cm'] = 0.0
        elif zero_rate > 0.80:
            zero_candidates = df.loc[ctrl_idx].loc[df.loc[ctrl_idx, 'tumor_size_cm'] == 0].index.to_numpy()
            need = int(np.ceil(zero_mask.sum() - 0.80 * len(ctrl_idx)))
            if zero_candidates.size > 0 and need > 0:
                flip = rng.choice(zero_candidates, size=min(need, zero_candidates.size), replace=False)
                df.loc[flip, 'tumor_size_cm'] = rng.uniform(1.0, 5.0, size=len(flip))

    df.loc[df['diagnosis_label'] == 1, 'tumor_size_cm'] = df.loc[df['diagnosis_label'] == 1, 'tumor_size_cm'].clip(0.5, 20.0)
    df.loc[df['diagnosis_label'] == 0, 'tumor_size_cm'] = df.loc[df['diagnosis_label'] == 0, 'tumor_size_cm'].clip(0.0, None)

    # ---------- 7) Hemodilution effect for controls: enforce negative Spearman(BMI, CA-125) ----------
    ctrl_idx = df.index[df['diagnosis_label'] == 0].to_numpy()
    if len(ctrl_idx) >= 3:
        delta_bmi = df.loc[ctrl_idx, 'bmi'] - 25.0
        hemo = np.exp(delta_bmi * np.log(0.97))
        hemo = np.clip(hemo, 0.75, 1.20)
        df.loc[ctrl_idx, 'ca125'] = (df.loc[ctrl_idx, 'ca125'] * hemo).clip(2.0, 2000.0)

        rho = df.loc[ctrl_idx, ['bmi', 'ca125']].corr(method='spearman').iloc[0, 1]
        if pd.isna(rho) or rho >= 0:
            sorted_idx = df.loc[ctrl_idx].sort_values('bmi', ascending=False).index.to_numpy()
            take = max(1, int(0.20 * len(sorted_idx)))
            sel = sorted_idx[:take]
            df.loc[sel, 'ca125'] = (df.loc[sel, 'ca125'] * rng.uniform(0.85, 0.95, size=len(sel))).clip(2.0, 2000.0)

    # ---------- 8) Menopause / MHT rate adjustments ----------
    mask55 = df['age'] >= 55
    if mask55.any():
        p_post55 = df.loc[mask55, 'menopausal_status'].mean()
        if p_post55 < 0.80:
            need = int(np.ceil(0.80 * mask55.sum() - df.loc[mask55, 'menopausal_status'].sum()))
            cand = df.loc[mask55 & (df['menopausal_status'] == 0)].index.to_numpy()
            if cand.size > 0 and need > 0:
                flip = rng.choice(cand, size=min(need, cand.size), replace=False)
                df.loc[flip, 'menopausal_status'] = 1
        elif p_post55 > 0.90:
            cand = df.loc[mask55 & (df['menopausal_status'] == 1)].index.to_numpy()
            need = int(np.ceil(df.loc[mask55, 'menopausal_status'].sum() - 0.85 * mask55.sum()))
            if cand.size > 0 and need > 0:
                flip = rng.choice(cand, size=min(need, cand.size), replace=False)
                df.loc[flip, 'menopausal_status'] = 0

    post_ctrl_idx = df.index[(df['diagnosis_label'] == 0) & (df['menopausal_status'] == 1)].to_numpy()
    post_case_idx = df.index[(df['diagnosis_label'] == 1) & (df['menopausal_status'] == 1)].to_numpy()

    def enforce_prop(index_array, col, low, high, target):
        if len(index_array) == 0:
            return
        cur = df.loc[index_array, col].mean()
        if cur < low:
            zeros = df.loc[index_array][df.loc[index_array, col] == 0].index.to_numpy()
            need = int(np.ceil(target * len(index_array) - df.loc[index_array, col].sum()))
            if need > 0 and zeros.size > 0:
                flip = rng.choice(zeros, size=min(need, zeros.size), replace=False)
                df.loc[flip, col] = 1
        elif cur > high:
            ones = df.loc[index_array][df.loc[index_array, col] == 1].index.to_numpy()
            need = int(np.ceil(df.loc[index_array, col].sum() - target * len(index_array)))
            if need > 0 and ones.size > 0:
                flip = rng.choice(ones, size=min(need, ones.size), replace=False)
                df.loc[flip, col] = 0

    enforce_prop(post_ctrl_idx, 'mht_use', 0.15, 0.25, target=0.20)
    enforce_prop(post_case_idx, 'mht_use', 0.20, 0.30, target=0.25)

    # ---------- 9) Recompute ultrasound_risk_score and tune to satisfy within-case correlations ----------
    def recompute_and_tune_risk(local_df):
        eps = 1e-6
        log_ca = np.log(local_df['ca125'] / 35.0 + eps)
        alpha = -4.2
        gamma_size = 0.35
        gamma_ca = 0.40
        for _ in range(12):
            risk_linear = alpha + gamma_size * local_df['tumor_size_cm'] + gamma_ca * log_ca
            risk = 1.0 / (1.0 + np.exp(-risk_linear))
            mask_no_mass = (local_df['diagnosis_label'] == 0) & (local_df['tumor_size_cm'] == 0)
            if mask_no_mass.any():
                risk.loc[mask_no_mass] = np.minimum(risk.loc[mask_no_mass], 0.079)
            cases_df = local_df[local_df['diagnosis_label'] == 1]
            if len(cases_df) > 3:
                rho_sz = cases_df['tumor_size_cm'].corr(risk.loc[cases_df.index], method='spearman')
                rho_ca = cases_df['ca125'].corr(risk.loc[cases_df.index], method='spearman')
                ok_sz = pd.notna(rho_sz) and rho_sz > 0.5
                ok_ca = pd.notna(rho_ca) and (0.3 <= rho_ca <= 0.5)
                if ok_sz and ok_ca:
                    return np.clip(risk, 0.0, 0.99)
                if not ok_sz:
                    gamma_size *= 1.10
                if pd.notna(rho_ca):
                    if rho_ca < 0.30:
                        gamma_ca *= 1.10
                    elif rho_ca > 0.50:
                        gamma_ca *= 0.90
            else:
                return np.clip(risk, 0.0, 0.99)
        risk_linear = alpha + gamma_size * local_df['tumor_size_cm'] + gamma_ca * log_ca
        return np.clip(1.0 / (1.0 + np.exp(-risk_linear)), 0.0, 0.99)

    df['ultrasound_risk_score'] = recompute_and_tune_risk(df)

    no_mass_ctrl_idx = df.index[(df['diagnosis_label'] == 0) & (df['tumor_size_cm'] == 0)].to_numpy()
    if len(no_mass_ctrl_idx) >= 1:
        q95 = df.loc[no_mass_ctrl_idx, 'ultrasound_risk_score'].quantile(0.95)
        if q95 > 0.08:
            high_idx = df.loc[no_mass_ctrl_idx][df.loc[no_mass_ctrl_idx, 'ultrasound_risk_score'] > 0.08].index.to_numpy()
            if len(high_idx) > 0:
                df.loc[high_idx, 'ultrasound_risk_score'] = df.loc[high_idx, 'ultrasound_risk_score'].apply(lambda x: min(x, 0.079))

    # ---------- 10) Ensure correlations within-cases: corr(size, CA-125) > 0 ----------
    cases_idx = df.index[df['diagnosis_label'] == 1].to_numpy()
    if len(cases_idx) >= 3:
        rho_size_ca = df.loc[cases_idx, ['tumor_size_cm', 'ca125']].corr(method='spearman').iloc[0, 1]
        if pd.isna(rho_size_ca) or rho_size_ca <= 0.0:
            top_pct = int(np.ceil(0.20 * len(cases_idx)))
            sorted_cases = df.loc[cases_idx].sort_values('tumor_size_cm', ascending=False).index.to_numpy()
            top_idx = sorted_cases[:max(1, top_pct)]
            df.loc[top_idx, 'ca125'] = (df.loc[top_idx, 'ca125'] * rng.uniform(1.05, 1.25, size=len(top_idx))).clip(2.0, 2000.0)

    # ---------- 11) BRCA+ cases skew younger ----------
    brca_case_idx = df.index[(df['diagnosis_label'] == 1) & (df['brca_status'] == 1)].to_numpy()
    if len(brca_case_idx) >= 2:
        med_age_brca1 = df.loc[brca_case_idx, 'age'].median()
        med_age_brca0 = df.loc[df.index[(df['diagnosis_label'] == 1) & (df['brca_status'] == 0)], 'age'].median()
        if pd.notna(med_age_brca1) and pd.notna(med_age_brca0) and not (med_age_brca1 < med_age_brca0):
            shift_n = max(1, int(0.25 * len(brca_case_idx)))
            shift_idx = rng.choice(brca_case_idx, size=shift_n, replace=False)
            df.loc[shift_idx, 'age'] = np.clip(df.loc[shift_idx, 'age'] - rng.integers(4, 11, size=len(shift_idx)), 18, 90)

    # ---------- FINAL: strict control CA-125 enforcement (after all ca125 changes) ----------
    ctrl_idx = df.index[df['diagnosis_label'] == 0].to_numpy()
    if len(ctrl_idx) >= 1:
        n_ctrl = len(ctrl_idx)
        target_min = int(np.ceil(0.01 * n_ctrl))
        target_max = int(np.floor(0.03 * n_ctrl))
        if target_max < target_min:
            target = int(round(0.02 * n_ctrl))
            target_min = target_max = max(0, target)

        # compute current high count
        current_high_idx = df.loc[ctrl_idx][df.loc[ctrl_idx, 'ca125'] > 35].index.to_numpy()
        current_high = len(current_high_idx)

        # If too many, deterministically lower highest ca125 values among controls until within bounds
        if current_high > target_max:
            excess = current_high - target_max
            # sort current_high_idx by ca125 descending - lower the largest ones first
            ordered = df.loc[current_high_idx, 'ca125'].sort_values(ascending=False).index.to_numpy()
            to_lower = ordered[:excess]
            # lower them into [20,34] but preserve other stats by using RNG
            df.loc[to_lower, 'ca125'] = rng.uniform(20.0, 34.0, size=len(to_lower))

        # If too few, raise lowest non-high controls up to >=36 until target_min reached
        elif current_high < target_min:
            deficit = target_min - current_high
            low_candidates = df.loc[ctrl_idx][df.loc[ctrl_idx, 'ca125'] <= 35].index.to_numpy()
            if len(low_candidates) > 0:
                # choose lowest CA125 among candidates first
                ordered_low = df.loc[low_candidates, 'ca125'].sort_values().index.to_numpy()
                to_raise = ordered_low[:min(deficit, len(ordered_low))]
                df.loc[to_raise, 'ca125'] = np.maximum(df.loc[to_raise, 'ca125'], rng.uniform(36.0, 60.0, size=len(to_raise)))

        # final safety check: if still out of bounds (rare due to rounding), force exact integer by further lowering or raising
        current_high_idx = df.loc[ctrl_idx][df.loc[ctrl_idx, 'ca125'] > 35].index.to_numpy()
        current_high = len(current_high_idx)
        if current_high > target_max:
            ordered = df.loc[current_high_idx, 'ca125'].sort_values(ascending=False).index.to_numpy()
            to_lower = ordered[:current_high - target_max]
            df.loc[to_lower, 'ca125'] = 34.0
        elif current_high < target_min:
            low_candidates = df.loc[ctrl_idx][df.loc[ctrl_idx, 'ca125'] <= 35].index.to_numpy()
            ordered_low = df.loc[low_candidates, 'ca125'].sort_values().index.to_numpy()
            to_raise = ordered_low[:(target_min - current_high)]
            df.loc[to_raise, 'ca125'] = 36.0

    # ---------- 12) Final hard clipping & type cast ----------
    df['age'] = df['age'].clip(18, 90).astype(int)
    df['bmi'] = df['bmi'].clip(16.0, 45.0).round(2)
    df['ca125'] = df['ca125'].clip(2.0, 2000.0)
    df.loc[df['diagnosis_label'] == 1, 'tumor_size_cm'] = df.loc[df['diagnosis_label'] == 1, 'tumor_size_cm'].clip(0.5, 20.0)
    df.loc[df['diagnosis_label'] == 0, 'tumor_size_cm'] = df.loc[df['diagnosis_label'] == 0, 'tumor_size_cm'].clip(0.0, None)
    df['ultrasound_risk_score'] = df['ultrasound_risk_score'].clip(0.0, 0.99).round(3)

    for col in ['menopausal_status', 'family_history', 'mht_use', 'brca_status', 'diagnosis_label', 'parity']:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)

    return df

post_synthetic_data = enforce_constraints(synthetic_data, seed=42)
post_synthetic_data.to_csv("data/final_ctGAN_database.csv", index=False)
print(f"Synthetic data saved to data/final_ctGAN_database.csv")
```

#### Plausibility checks (All Passed)

```python
import numpy as np
import pandas as pd


df=pd.read_csv("data/final_ctGAN_database.csv")

def pct(x):
    try: return 100*float(x)
    except: return np.nan

def report(desc, ok, val):
    status = "PASS" if ok else "FAIL"
    # Pretty-print tuples and scalars
    if isinstance(val, tuple):
        v = ", ".join(f"{x:.3f}" if isinstance(x, (int,float,np.floating)) else str(x) for x in val)
    elif isinstance(val, (int,float,np.floating)):
        v = f"{val:.3f}"
    else:
        v = str(val)
    print(f"[{status}] {desc:<60} -> {v}")

controls = df[df["diagnosis_label"] == 0]
cases    = df[df["diagnosis_label"] == 1]

checks = []

# --- Cohort composition ---
prev = df["diagnosis_label"].mean()
checks.append(("Cancer fraction ≥ 30%", prev >= 0.30, prev))

# --- BRCA prevalence ---
brca_case_rate = cases["brca_status"].mean() if len(cases) else np.nan
checks.append(("BRCA+ among cases in [12%, 18%]", 0.12 <= brca_case_rate <= 0.18, brca_case_rate))

# --- CA-125 behavior ---
healthy_high35 = (controls["ca125"] > 35).mean() if len(controls) else np.nan
case_high200   = (cases["ca125"] > 200).mean() if len(cases) else np.nan
checks.append(("Healthy CA-125 >35 U/mL ≈ 1–3%", 0.01 <= healthy_high35 <= 0.03, healthy_high35))
checks.append(("Cancer CA-125 >200 U/mL ≥ 35%", case_high200 >= 0.35, case_high200))

# --- Tumor size ---
median_size_cases = cases["tumor_size_cm"].median() if len(cases) else np.nan
nz_ctrl = controls.loc[controls["tumor_size_cm"] > 0, "tumor_size_cm"]
median_size_ctrl_nonzero = nz_ctrl.median() if len(nz_ctrl) else np.nan
checks.append(("Median tumor size (cases) in [8, 11] cm", 8.0 <= median_size_cases <= 11.0, median_size_cases))
if pd.notna(median_size_ctrl_nonzero):
    checks.append(("Median benign cyst (controls) in [2, 4] cm", 2.0 <= median_size_ctrl_nonzero <= 4.0, median_size_ctrl_nonzero))

# Zero-mass proportion in controls (expect high; ~60–80%)
zero_mass_rate = (controls["tumor_size_cm"] == 0).mean() if len(controls) else np.nan
checks.append(("Controls: zero-mass proportion in [0.6, 0.8]", 0.60 <= zero_mass_rate <= 0.80, zero_mass_rate))

# --- Ultrasound risk sanity ---
mean_risk_cases = cases["ultrasound_risk_score"].mean() if len(cases) else np.nan
mean_risk_ctrls = controls["ultrasound_risk_score"].mean() if len(controls) else np.nan
checks.append(("Mean risk (cases) > Mean risk (controls)", mean_risk_cases > mean_risk_ctrls, (mean_risk_cases, mean_risk_ctrls)))

# Controls w/ no mass should be very low risk (use 95th percentile to avoid mean skew)
no_mass = controls.loc[controls["tumor_size_cm"] == 0, "ultrasound_risk_score"]
q95 = no_mass.quantile(0.95) if len(no_mass) else np.nan
checks.append(("Controls no-mass: 95th percentile risk ≤ 0.08", q95 <= 0.08, q95))

# --- Menopause / MHT checks ---
p_post_55 = (df.loc[df["age"] >= 55, "menopausal_status"].mean()
             if (df["age"] >= 55).any() else np.nan)
checks.append(("P(postmenopause | age ≥ 55) in [0.80, 0.90]", 0.80 <= p_post_55 <= 0.90, p_post_55))

mht_post_ctrl = controls.loc[controls["menopausal_status"] == 1, "mht_use"].mean() if (controls["menopausal_status"] == 1).any() else np.nan
mht_post_case = cases.loc[cases["menopausal_status"] == 1, "mht_use"].mean() if (cases["menopausal_status"] == 1).any() else np.nan
checks.append(("MHT among post-menopause (controls) ≈ 0.20±0.05", 0.15 <= mht_post_ctrl <= 0.25, mht_post_ctrl))
checks.append(("MHT among post-menopause (cases) ≈ 0.25±0.05", 0.20 <= mht_post_case <= 0.30, mht_post_case))

# --- Hemodilution sanity (controls only): BMI vs CA-125 should be negative ---
rho_bmi_ca125_ctrl = controls[["bmi","ca125"]].corr(method="spearman").iloc[0,1] if len(controls) else np.nan
checks.append(("Controls: Spearman(BMI, CA-125) < 0", rho_bmi_ca125_ctrl < 0, rho_bmi_ca125_ctrl))

# --- Within-cases correlations (intended behavior) ---
def safe_corr(df_sub, a, b, method="spearman"):
    if len(df_sub) < 3 or df_sub[[a,b]].dropna().shape[0] < 3:
        return np.nan
    return df_sub[[a,b]].corr(method=method).iloc[0,1]

rho_size_risk_case  = safe_corr(cases, "tumor_size_cm", "ultrasound_risk_score")
rho_ca_risk_case    = safe_corr(cases, "ca125", "ultrasound_risk_score")
rho_size_ca_case    = safe_corr(cases, "tumor_size_cm", "ca125")

checks.append(("Cases: corr(size, risk) > 0.5", rho_size_risk_case > 0.5, rho_size_risk_case))
checks.append(("Cases: corr(CA-125, risk) in [0.3, 0.5]", 0.3 <= rho_ca_risk_case <= 0.5, rho_ca_risk_case))
checks.append(("Cases: corr(size, CA-125) > 0", rho_size_ca_case > 0.0, rho_size_ca_case))

# --- BRCA & age relationship (you already had this; keeping it) ---
med_age_brca1 = cases.loc[cases["brca_status"]==1, "age"].median() if (cases["brca_status"]==1).any() else np.nan
med_age_brca0 = cases.loc[cases["brca_status"]==0, "age"].median() if (cases["brca_status"]==0).any() else np.nan
checks.append(("BRCA+ cases skew younger (median)", med_age_brca1 < med_age_brca0, (med_age_brca1, med_age_brca0)))

# --- Hard bounds sanity (no impossible values) ---
ok_bounds = (
    (df["bmi"].between(16, 45).all()) and
    (df["age"].between(18, 90).all()) and
    (df["ca125"].between(2, 2000).all()) and
    (df.loc[df["diagnosis_label"]==1, "tumor_size_cm"].between(0.5, 20).all()) and
    (df.loc[df["diagnosis_label"]==0, "tumor_size_cm"].ge(0).all())
)
checks.append(("All features within specified bounds", ok_bounds, ok_bounds))

# --- Print report ---
print("Plausibility checks:")
for desc, ok, val in checks:
    report(desc, ok, val)
```
